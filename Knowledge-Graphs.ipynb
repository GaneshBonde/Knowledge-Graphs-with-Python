```python
import pandas as pd
%matplotlib inline
```

# 1. Exploratory Data Analysis (EDA)

# 1.1 Load the data

```python
# import the data from github
movie_plots_data = pd.read_csv("https://raw.githubusercontent.com/GaneshBonde/Knowledge-Graphs-with-Python/main/dataset/wiki_movie_plots_deduped.csv", 
                              error_bad_lines=False, delimiter=',', skiprows=0, low_memory=False)
```

```python
movie_plots_data.shape
```

# 1.2 Inspect Raw Data

```python
movie_plots_data.head()
```

# 1.3 Visualize Country Of Origin

```python
movie_plots_data.groupby(['Origin/Ethnicity']).size().sort_values(ascending=True).plot.barh()
```

# 1.4 Visualize Release Year Trend

```python
movie_plots_data.groupby(['Release Year']).size().plot(kind='bar',figsize=(22,4), grid = True )
```

# 1.5 Visualize Genre Breakout

```python
#movie_plots_data[movie_plots_data['Genre']!='unknown'].groupby(['Genre']).size()
#.sort_values(ascending=True).tail(25).plot.barh(figsize=(4,8), grid=True)
movie_plots_data[movie_plots_data['Genre']!='unknown'].groupby(['Genre']).size().sort_values(ascending=True).tail(25).plot.barh(figsize=(4,8), grid=True)
```

# 1.6 Visualize Top Director

```python
#movie_plots_data[movie_plots_data['Director']!='Unknown'].groupby(['Director'])
#.size().sort_values(ascending=True).tail(20).plot.barh(figsize=(4,8), grid=True)
movie_plots_data[movie_plots_data['Director']!='Unknown'].groupby(['Director']).size().sort_values(ascending=True).tail(20).plot.barh(figsize=(4,8), grid=True)
```

# 1.7 Visualize Top Titles

```python
#movie_plots_data[movie_plots_data['Title']!='Unknown'].groupby(['Title']).size()
#.sort_values(ascending=True).tail(20).plot.barh(figsize=(4,8),grid=True)
movie_plots_data[movie_plots_data['Title']!='Unknown'].groupby(['Title']).size().sort_values(ascending=True).tail(20).plot.barh(figsize=(4,8),grid=True)
```

# 1.8 Visualize Top Directors

```python
#movie_plots_data[movie_plots_data['Director']!='Unknown'].groupby(['Director']).size()
#.sort_values(ascending=True).tail(20).plot.barh(figsize=(4,8),grid=True)
movie_plots_data[movie_plots_data['Director']!='Unknown'].groupby(['Director']).size().sort_values(ascending=True).tail(20).plot.barh(figsize=(4,8),grid=True)
```

# 1.9 Visualize Top Casts

```python
#movie_plots_data[movie_plots_data['Cast']!='Unknown'].groupby(['Cast']).size()
#.sort_values(ascending=True).tail(20).plot.barh(figsize=(4,8),grid=True)
movie_plots_data[movie_plots_data['Cast']!='Unknown'].groupby(['Cast']).size().sort_values(ascending=True).tail(20).plot.barh(figsize=(4,8),grid=True)
```
